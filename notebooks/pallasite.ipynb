```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/home/eric_baldwin/ddiMain/midterm/Meteor_Landing_Data/data/Meteorite_Landings_Cleaned.csv'
df = pd.read_csv(file_path)
```

```python
df_pallasite = df[(df['recclass'] == 'Pallasite')]

# Display the first few rows to verify the data
df_pallasite.head(30)
```

```python
recclass_counts = df_pallasite['recclass'].value_counts()
print(recclass_counts)
```

```python
# Filter the DataFrame to only include rows where recclass is 'Pallasite' and GeoLocation is not (0.0, 0.0)
df_filtered = df[(df['recclass'] == 'Pallasite') & (df['GeoLocation'] != '(0.0, 0.0)')]

# Display the first few rows to verify the data
df_filtered.head()
```

```python
# Calculate the average mass (g) for Pallasite
average_mass_pallasite = df_filtered['mass (g)'].mean()

# Display the average mass
print(f"Average mass (g) of Pallasite meteorites: {average_mass_pallasite}")
```

```python



# Plotting functions
def plot_bar_chart(df):
    """Creates and displays a bar chart of the mass distribution."""
    plt.figure(figsize=(12, 8))
    sns.histplot(df['mass (g)'], bins=30, kde=False)
    plt.title('Mass Distribution of Pallasite Meteorites')
    plt.xlabel('Mass (g)')
    plt.ylabel('Number of Rows')
    plt.show()

def plot_box_chart(df):
    """Creates and displays a box plot of the mass (g)."""
    plt.figure(figsize=(12, 8))
    sns.boxplot(y=df['mass (g)'])
    plt.title('Box Plot of Pallasite Meteorite Mass')
    plt.ylabel('Mass (g)')
    plt.show()

def plot_violin_chart(df):
    """Creates and displays a violin plot of the mass (g)."""
    plt.figure(figsize=(12, 8))
    sns.violinplot(y=df['mass (g)'])
    plt.title('Violin Plot of Pallasite Meteorite Mass')
    plt.ylabel('Mass (g)')
    plt.show()

def plot_histogram(df):
    """Creates and displays a histogram of the mass (g)."""
    plt.figure(figsize=(12, 8))
    sns.histplot(df['mass (g)'], bins=30, kde=True)
    plt.title('Histogram of Pallasite Meteorite Mass')
    plt.xlabel('Mass (g)')
    plt.ylabel('Frequency')
    plt.show()

# Generate and display the charts
plot_bar_chart(df_filtered)
plot_box_chart(df_filtered)
plot_violin_chart(df_filtered)
plot_histogram(df_filtered)
```
