```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = '/home/eric_baldwin/ddiMain/midterm/Meteor_Landing_Data/data/Meteorite_Landings_Cleaned.csv'
df = pd.read_csv(file_path)
```

```python
table_shape = df.shape
table_shape
```

```python
# Group by year and count the number of meteorites found each year
meteorites_per_year = df['year'].value_counts().sort_index()
meteorites_per_year.head()
```

```python
# Plot the data
plt.figure(figsize=(12, 6))
sns.lineplot(x=meteorites_per_year.index, y=meteorites_per_year.values)
plt.title('Number of Meteorites Found Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Meteorites Found')
plt.grid(True)
plt.show()
```

```python
top_3_recclass = [
    'L6',
    'H5',
    'L5'
]

df = df[df['year'] >= 1970.0]

# Filter the DataFrame to only include rows with the top 5 recclass
df_top_3 = df[df['recclass'].isin(top_3_recclass)]


# Plot individual histograms for the top 3 recclass
plt.figure(figsize=(15, 15))

for i, recclass in enumerate(top_3_recclass, 1):
    plt.subplot(3, 1, i)  # Create a subplot for each histogram
    sns.histplot(df_top_3[df_top_3['recclass'] == recclass]['year'], bins=30)
    plt.title(f'Histogram of {recclass} Found Over the Years (1970 - 2012)')
    plt.xlabel('Year')
    plt.ylabel('Number of Meteorites Found')
    plt.ylim(0, 1000)  # Set y-axis limit to show 0 - 1000
    plt.xlim(1970, 2012)  # Set x-axis limit to show 1970 - 2010
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
top_3_recclass = [
    'L6',
    'H5',
    'L5'
]

df = df[df['year'] >= 1970.0]

# Filter the DataFrame to only include rows with the top 5 recclass
df_top_3 = df[df['recclass'].isin(top_3_recclass)]

# Plot the KDE plots for the top 5 recclass
plt.figure(figsize=(12, 8))
for recclass in top_3_recclass:
    sns.kdeplot(df_top_3[df_top_3['recclass'] == recclass]['year'], label=recclass, fill=True)
plt.xticks(range(int(df['year'].min()), int(df['year'].max()) + 1, 5))
plt.title('Top 3 Meteorite Types Found Over the Years (1970 - 2012)')
plt.xlabel('Year')
plt.ylabel("Meteorites Found")
plt.legend()
plt.grid(True)
plt.show()
```
