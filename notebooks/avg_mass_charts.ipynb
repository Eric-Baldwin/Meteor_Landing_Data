```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Ensure plots are displayed inline in the notebook
%matplotlib inline

# Load the data
file_path = '/home/eric_baldwin/ddiMain/midterm/Meteor_Landing_Data/data/Meteorite_Landings_Cleaned.csv'
df = pd.read_csv(file_path)
```

```python
# Determine the number of rows with mass <= X
rows_with_mass_less_equal_X = df[df['mass (g)'] <= 125].shape[0]

# Display the result
print(f"Rows: {rows_with_mass_less_equal_X}")
```

```python

# Calculate the average mass (g)
average_mass = df['mass (g)'].mean()

# Display the average mass
print(f"Average mass (g) of meteorites: {average_mass}")
```

```python
df = df[df['mass (g)'] <= 1000]
```

```python

def plot_box_chart(df):
    """Creates and displays a box plot of the mass (g)."""
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df['mass (g)'])
    plt.title('Box Plot of Meteorite Mass')
    plt.ylabel('Mass (g)')
    plt.show()
plot_box_chart(df)
```

```python

# Plotting functions
def plot_bar_chart(df):
    """Creates and displays a bar chart of the mass distribution."""
    plt.figure(figsize=(12, 4))
    sns.histplot(df['mass (g)'], bins=100, kde=False)
    plt.title('Mass Distribution of Meteorites')
    plt.xlabel('Mass (g)')
    plt.ylabel('Number of Meteorites Found')
    plt.show()
plot_bar_chart(df)
```

```python

def plot_line_chart(avg_mass):
    """Creates and displays a line chart of the average mass."""
    plt.figure(figsize=(10, 6))
    plt.plot(['Average Mass'], [avg_mass], marker='o')
    plt.title('Average Mass of Meteorites')
    plt.ylabel('Mass (g)')
    plt.xlabel('')
    plt.show()
plot_line_chart(average_mass)
```

```python

def plot_violin_chart(df):
    """Creates and displays a violin plot of the mass (g)."""
    plt.figure(figsize=(10, 6))
    sns.violinplot(y=df['mass (g)'])
    plt.title('Violin Plot of Meteorite Mass')
    plt.ylabel('Mass (g)')
    plt.show()
plot_violin_chart(df)
```

```python

def plot_histogram(df):
    """Creates and displays a histogram of the mass (g)."""
    plt.figure(figsize=(10, 6))
    sns.histplot(df['mass (g)'], bins=30, kde=True)
    plt.title('Histogram of Meteorite Mass')
    plt.xlabel('Mass (g)')
    plt.ylabel('Frequency')
    plt.show()
plot_histogram(df)
```
